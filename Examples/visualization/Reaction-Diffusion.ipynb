```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from datetime import date
def today(): return date.today().strftime("%Y%m%d")

import scipy.io as io

from os.path import join as join_path
FIGURE_EXPORT_PATH = "./"
```

```python
def check_percent(bic_scores, complexities):
    slope = (bic_scores[1:]-bic_scores[:-1])/(complexities[1:]-complexities[:-1])
    slope_index = np.argmin(slope)
    percent_improve = 100*np.abs(bic_scores[slope_index+1]-bic_scores[slope_index])/np.abs(bic_scores[slope_index])
    percent_from_1 = 100*np.abs(bic_scores[slope_index+1]-bic_scores[0])/np.abs(bic_scores[0])
    return slope_index+1, percent_improve, percent_from_1

def percent_coeff(pred, ground):
    return 100*(np.abs(pred-ground)/np.abs(ground))
```

## 2D

```python
udata = io.loadmat("./reaction_diffustion_2d_u_stlsq_bnb_frols_noise_lv2.mat")
rel_bic_scores_u = udata["rel_bic_scores"][0]
complexities_u = udata["complexities"][0]

vdata = io.loadmat("./reaction_diffustion_2d_v_stlsq_bnb_frols_noise_lv2.mat")
rel_bic_scores_v = vdata["rel_bic_scores"][0]
complexities_v = vdata["complexities"][0]
```

```python
mini = min(np.minimum(complexities_u, complexities_v))
maxi = max(np.maximum(complexities_u, complexities_v))
```

```python
print(check_percent(rel_bic_scores_u*1e3+udata["min_bic_score"][0], complexities_u))
print(check_percent(rel_bic_scores_v*1e3+vdata["min_bic_score"][0], complexities_v))
```

```python
with plt.style.context('science'):
    fig, ax = plt.subplots()
    ax.plot([], [], ' ', label="Noise level 2\%")
    ax.plot([], [], ' ', label="Predicted vars.")
    ax.plot(complexities_u[:6], rel_bic_scores_u[:6], color='green', linewidth=2.5//2, label='$u_t(x, y, t)$', zorder=1)
    ax.plot(complexities_u[5:7], rel_bic_scores_u[5:7], '--', color='green', linewidth=2.5//2, zorder=1)
    ax.plot(complexities_u[6:], rel_bic_scores_u[6:], color='green', linewidth=2.5//2, zorder=1)
    ax.scatter(complexities_u, rel_bic_scores_u, marker='o', fc='none', ec='green', s=25, zorder=1)
    
    ax.plot(complexities_v[:6], rel_bic_scores_v[:6], color='red', linewidth=2.5//2, label='$v_t(x, y, t)$', zorder=2)
    ax.plot(complexities_v[5:7], rel_bic_scores_v[5:7], '--', color='red', linewidth=2.5//2, zorder=2)
    ax.plot(complexities_v[6:], rel_bic_scores_v[6:], color='red', linewidth=2.5//2, zorder=2)
    ax.scatter(complexities_v, rel_bic_scores_v, marker='o', fc='none', ec='red', s=25, zorder=2)
    
    plt.xlim(mini-0.3, maxi+0.3)
    plt.xticks([i for i in range(mini, maxi+1)])
    plt.ylabel(r"Relative BIC ($\times 10^{3}$)")
    plt.xlabel("Complexity")
    plt.tight_layout()
    plt.legend()
    # plt.savefig("./2D.pdf")
    plt.show()
```
