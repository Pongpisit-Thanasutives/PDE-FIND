```python
%matplotlib inline
import matplotlib.pyplot as plt

import sys; sys.path.append('../')
from function_collections import colvec
from misc import h5file
import numpy as np

import scipy.io as sio
import sympy
from sympy import Symbol, parse_expr, sympify, lambdify, Lambda

import pandas as pd
import subprocess

from scipy.optimize import curve_fit

MAIN_SEED = 1234
```

```python
un = np.load("./denoised_files/KdV_rep_sine_noise30_reduced_dctV2ksvdreg_25x25.npy")
```

```python
x = colvec(sio.loadmat("../Datasets/KdV_sine_rep_big.mat")['x'])
```

```python
from pysr import PySRRegressor

deterministic = True

if not deterministic:
    model = PySRRegressor(
        niterations=200,  # < Increase me for better results
        binary_operators=["+", "*"],
        unary_operators=["sin", "cos"], 
        deterministic=deterministic
    )
else:
    np.random.seed(MAIN_SEED)
    model = PySRRegressor(
        niterations=200,  # < Increase me for better results
        binary_operators=["+", "*"],
        unary_operators=["sin", "cos"], 
        # For a deterministic result
        deterministic=deterministic, 
        random_state=MAIN_SEED, 
        procs=0, 
        multithreading=False
    )
```

```python
model.fit(x, un[:,0])
```

```python
model.get_hof().to_csv("./hall/hall.csv", index=None)
```
