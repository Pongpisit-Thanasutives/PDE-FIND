```python
%matplotlib inline
import matplotlib.pyplot as plt

import sys; sys.path.append('../')
from misc import h5file
import numpy as np

import scipy.io as sio
import sympy
from sympy import Symbol, parse_expr, sympify, lambdify, Lambda

import pandas as pd
import subprocess

from scipy.optimize import curve_fit

MAIN_SEED = 1234
```

```python
un = np.load("../../UBIC-play/Examples/denoised_files/burgers_shock_dictlearn_denoised30_components32_8x8.npy")
data = sio.loadmat("../../UBIC-play/Datasets/burgers_shock.mat")
x = data['x']
```

```python
from pysr import PySRRegressor

deterministic = True

if not deterministic:
    model = PySRRegressor(
        niterations=50,  # < Increase me for better results
        binary_operators=["+", "*"],
        unary_operators=["exp", "square"], 
        deterministic=deterministic
    )
else:
    np.random.seed(MAIN_SEED)
    model = PySRRegressor(
        niterations=50,  # < Increase me for better results
        binary_operators=["+", "*"],
        unary_operators=["sin", "cos"], 
        # For a deterministic result
        deterministic=deterministic, 
        random_state=MAIN_SEED, 
        procs=0, 
        multithreading=False
    )
```

```python
model.fit(x, un[:,0])
```

```python
save_to = "./hof.csv"
model.get_hof().to_csv(save_to, index=None)
```
