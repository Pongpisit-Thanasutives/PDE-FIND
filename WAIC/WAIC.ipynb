```python
%matplotlib inline
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
dataset_name = "Burgers"
ic_collections = json.load(open(f"/Users/pongpisit/Desktop/fun/R/data/{dataset_name}/ic_collections/ic_collections_waic_cv.json"))
complexities = np.arange(len(ic_collections["cvs"]))+1

with plt.style.context(['science']):
    fig, ax = plt.subplots()
    ax.plot(complexities, ic_collections["waic"], 'o-', c='red', markerfacecolor='none', label = 'WAIC (w/o $\lambda_{\\textrm{U}}\\textrm{CV}^{k}$)')
    ax.plot(complexities, ic_collections["waic_cvs"], 'o-', c='black', markerfacecolor='none', label = 'UWAIC (w/ $\lambda_{\\textrm{U}}\\textrm{CV}^{k}$)')
    ax.plot([], [], linestyle='', marker='', label='($\lambda_{\\textrm{U}} = 1$)')

    ax.annotate('min', fontsize=16, color='red',
                 xy=(complexities[np.argmin(ic_collections["waic"])], ic_collections["waic"][np.argmin(ic_collections["waic"])]), 
                 xytext=(complexities[np.argmin(ic_collections["waic"])], ic_collections["waic"][np.argmin(ic_collections["waic"])]+2000), 
                 arrowprops={'arrowstyle': '->', 'color':'red'})

    ax.annotate('min\checkmark', fontsize=16, color='black',
                 xy=(complexities[np.argmin(ic_collections["waic_cvs"])], ic_collections["waic_cvs"][np.argmin(ic_collections["waic_cvs"])]), 
                 xytext=(complexities[np.argmin(ic_collections["waic_cvs"])], ic_collections["waic_cvs"][np.argmin(ic_collections["waic_cvs"])]+2000), 
                 arrowprops={'arrowstyle': '->', 'color':'black'})

    ax.set_xticks(complexities)
    ax.set_ylabel("$\\textrm{UWAIC}$", fontsize=12)
    ax.set_xlabel("Support sizes ($s_{k}$)", fontsize=12)
    plt.legend(fontsize=12)
    plt.savefig("../Figures/Burgers/burgers_waic.pdf")
    plt.show()
```
