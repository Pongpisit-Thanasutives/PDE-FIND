```python
def vis(rec):
    global x, t
    plt.contourf(*np.meshgrid(x, t), rec, levels=10)
    plt.colorbar()
    plt.show()
```

```python
data = loadmat('../Datasets/kuramoto_sivishinky.mat')
u = data['uu']
x = data['x'][:,0]
t = data['tt'][0,:]
dt = t[1]-t[0]
dx = x[2]-x[1]

np.random.seed(0)
noise_lv = 30
noise = 0.01*np.abs(noise_lv)*(u.std())*np.random.randn(u.shape[0], u.shape[1])
un = u+noise
```

```python
((un-u)**2).mean()
```

```python
dysmho = DySMHO(un.T, div=10)
denoised = dysmho.smooth(window_size=1) # 1(0.01047), 3(0.01036), 5(0.01025)
((denoised.T-u)**2).mean(), dysmho.best_window_size
```

```python
reduced_un = denoised.T
```
